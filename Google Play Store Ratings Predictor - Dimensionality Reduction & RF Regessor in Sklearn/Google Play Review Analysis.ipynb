```python
import pandas as pd
import numpy as np
```

```python
review_df = pd.read_csv('Processed Google Play Reviews.csv')
```

```python
store_df = pd.read_csv('processed play_store data.csv')
```

```python
review_df.dtypes
```

```python
store_df['App'] = store_df.App.astype(str)
review_df['App'] = review_df.App.astype(str)
```

```python
df = review_df.merge(store_df[['App', 'Rating']], on='App', how='left')
```

```python
df = df.drop_duplicates(subset='App')
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(how='any')
df = df.reset_index(drop=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
y = df[['Rating']]
```

```python
X = df.drop(columns=['Rating', 'App'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = RandomForestRegressor(random_state=42)
```

```python
model.fit(X_train, np.ravel(y_train))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
